```python
df["TransactionDate"] = pd.to_datetime(df["TransactionDate"])
df["TransactionDate"].dt.year.value_counts().plot(kind='bar', figsize=(10, 5), color='purple')
```

```python
df["PreviousTransactionDate"] =pd.to_datetime(df["PreviousTransactionDate"])
df["PreviousTransactionDate"].dt.year.value_counts().plot(kind='bar', figsize=(10, 5), color='orange')
```

```python
df["Transaction_Duration"] = (
    df['TransactionDate'] - df['PreviousTransactionDate']).dt.total_seconds() / 3600
```

```python
df = df.drop(columns=["TransactionID",'AccountID','DeviceID','IP Address',"PreviousTransactionDate","TransactionDate"])
df
```

# Correlation matrix

```python
# Encoding
```

```python
cat_cols = ['TransactionType', 'Location', 'MerchantID', 'Channel', 'CustomerOccupation']
df = pd.get_dummies(df, columns=cat_cols, drop_first=True)
```

```python
df
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
df_scaled = pd.DataFrame(df_scaled, columns=df.columns)
```

```python
df_scaled
```

# Isolation Forest for anomaly detection

```python
from sklearn.ensemble import IsolationForest
```

```python
iso_forest = IsolationForest(n_estimators=100, contamination=0.01, random_state=42)
df['anomaly_score_iso'] = iso_forest.fit_predict(df)
df['is_anomaly_iso'] = df['anomaly_score_iso'].apply(lambda x: 1 if x == -1 else 0)
```

```python
anomalies = df[df['is_anomaly_iso'] == 1]
print("Detected Anomalies (Isolation Forest):")
print(anomalies[['TransactionAmount', 'AccountBalance', 'Transaction_Duration', 'is_anomaly_iso']].head())
```

# Anomaly Detection - Isolation Forest

```python
sns.scatterplot(x='TransactionAmount', y='AccountBalance', hue='is_anomaly_iso', data=df)
plt.title("Anomaly Detection - Isolation Forest")
plt.xlabel("Transaction Amount")
plt.ylabel("Account Balance")
plt.legend(title="Is Anomaly")
plt.tight_layout()
plt.show()
```

# One Class SVM

```python
from sklearn.svm import OneClassSVM
oc_svm = OneClassSVM(kernel='rbf', nu=0.01, gamma='scale')
df['anomaly_score_svm'] = oc_svm.fit_predict(df_scaled)
df['is_anomaly_svm'] = df['anomaly_score_svm'].apply(lambda x: 1 if x == -1 else 0)
```

# Anomaly Detection - One-Class SVM

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TransactionAmount', y='AccountBalance', hue='is_anomaly_svm', data=df)
plt.title("Anomaly Detection - One-Class SVM")
plt.xlabel("Transaction Amount")
plt.ylabel("Account Balance")
plt.legend(title="Is Anomaly")
plt.tight_layout()
plt.show()
```

# Combine the anomaly scores from both methods

```python
# Combine the anomaly decisions
df['is_anomaly_combined'] = ((df['is_anomaly_iso'] + df['is_anomaly_svm']) > 0).astype(int)
```

```python
# Compare counts
print("Isolation Forest:", df['is_anomaly_iso'].sum())
print("One-Class SVM:", df['is_anomaly_svm'].sum())
print("Combined:", df['is_anomaly_combined'].sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(x='TransactionAmount', y='AccountBalance', hue='is_anomaly_combined', data=df)
plt.title("Anomaly Detection - Combined Result")
plt.xlabel("Transaction Amount")
plt.ylabel("Account Balance")
plt.legend(title="Is Anomaly")
plt.tight_layout()
plt.show()
```
